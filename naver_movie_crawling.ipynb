```python
#실관람객 요약 데이터 크롤링 코드

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import pandas as pd

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument('--headless')  # GUI 없이 실행
chrome_options.add_argument('--no-sandbox')  # 샌드박스 모드 비활성화
chrome_options.add_argument('--disable-dev-shm-usage')  # /dev/shm 파티션 사용 안함

# 웹 드라이버 설정 및 초기화
driver = webdriver.Chrome(options=chrome_options)

# 영화 URL 목록
movie_urls = [
    {"name": "패션왕(2014)", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1842967&qvt=0&query=%EC%98%81%ED%99%94%20%ED%8C%A8%EC%85%98%EC%99%95%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "신과함께 죄와벌", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1819142&qvt=0&query=%EC%8B%A0%EA%B3%BC%ED%95%A8%EA%BB%98-%EC%A3%84%EC%99%80%20%EB%B2%8C%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "치즈인더트랩(2018)", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=4032538&qvt=0&query=%EC%98%81%ED%99%94%20%EC%B9%98%EC%A6%88%EC%9D%B8%EB%8D%94%ED%8A%B8%EB%9E%A9%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "여중생a", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=6368982&qvt=0&query=%EC%98%81%ED%99%94%20%EC%97%AC%EC%A4%91%EC%83%9DA%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "신과함께 인과연", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=6684567&qvt=0&query=%EC%8B%A0%EA%B3%BC%ED%95%A8%EA%BB%98-%EC%9D%B8%EA%B3%BC%20%EC%97%B0%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "아이들은 즐겁다", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=18110448&qvt=0&query=%EC%98%81%ED%99%94%20%EC%95%84%EC%9D%B4%EB%93%A4%EC%9D%80%20%EC%A6%90%EA%B2%81%EB%8B%A4%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "천박사", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=28822499&qvt=0&query=%EC%B2%9C%EB%B0%95%EC%82%AC%20%ED%87%B4%EB%A7%88%20%EC%97%B0%EA%B5%AC%EC%86%8C%3A%20%EC%84%A4%EA%B2%BD%EC%9D%98%20%EB%B9%84%EB%B0%80%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "모럴센스", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=17988950&qvt=0&query=%EC%98%81%ED%99%94%20%EB%AA%A8%EB%9F%B4%EC%84%BC%EC%8A%A4%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "그녀의 버킷리스트", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=26087184&qvt=0&query=%EC%98%81%ED%99%94%20%EA%B7%B8%EB%85%80%EC%9D%98%20%EB%B2%84%ED%82%B7%EB%A6%AC%EC%8A%A4%ED%8A%B8%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "반드시 잡는다", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=4146980&qvt=0&query=%EB%B0%98%EB%93%9C%EC%8B%9C%20%EC%9E%A1%EB%8A%94%EB%8B%A4%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "바보", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1773584&qvt=0&query=%EC%98%81%ED%99%94%20%EB%B0%94%EB%B3%B4%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "순정만화", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1783758&qvt=0&query=%EC%88%9C%EC%A0%95%EB%A7%8C%ED%99%94%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "이끼", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1790148&qvt=0&query=%EC%98%81%ED%99%94%20%EC%9D%B4%EB%81%BC%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "그대를 사랑합니다", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1807125&qvt=0&query=%EA%B7%B8%EB%8C%80%EB%A5%BC%EC%82%AC%EB%9E%91%ED%95%A9%EB%8B%88%EB%8B%A4%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "이웃사람", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1826497&qvt=0&query=%EC%98%81%ED%99%94%EC%9D%B4%EC%9B%83%EC%82%AC%EB%9E%8C%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "26년", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1785250&qvt=0&query=%EC%98%81%ED%99%94%2026%EB%85%84%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "전설의주먹", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1823159&qvt=0&query=%EC%98%81%ED%99%94%20%EC%A0%84%EC%84%A4%EC%9D%98%20%EC%A3%BC%EB%A8%B9%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "은밀하게 위대하게", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1825983&qvt=0&query=%EC%98%81%ED%99%94%20%EC%9D%80%EB%B0%80%ED%95%98%EA%B2%8C%20%EC%9C%84%EB%8C%80%ED%95%98%EA%B2%8C%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "강철비", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=4348174&qvt=0&query=%EA%B0%95%EC%B2%A0%EB%B9%84%201%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "0.0MHz", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=8603029&qvt=0&query=%EC%98%81%ED%99%94%200.0MHz%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "발광하는 현대사", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=9832197&qvt=0&query=%EB%B0%9C%EA%B4%91%ED%95%98%EB%8A%94%20%ED%98%84%EB%8C%80%EC%82%AC%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "시동", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=9503606&qvt=0&query=%EC%98%81%ED%99%94%20%EC%8B%9C%EB%8F%99%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "해치지않아", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=8807396&qvt=0&query=%EC%98%81%ED%99%94%20%ED%95%B4%EC%B9%98%EC%A7%80%EC%95%8A%EC%95%84%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "강철비2", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=10080109&qvt=0&query=%EA%B0%95%EC%B2%A0%EB%B9%842%3A%20%EC%A0%95%EC%83%81%ED%9A%8C%EB%8B%B4%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "샤크:더 비기닝", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=18642550&qvt=0&query=%EC%83%A4%ED%81%AC%20%3A%20%EB%8D%94%20%EB%B9%84%EA%B8%B0%EB%8B%9D%20%EA%B4%80%EB%9E%8C%ED%8F%89"}
]


def fetch_data_from_tab(movie):
    driver.get(movie['url'])
    time.sleep(2)  # 웹 페이지 로드 대기

    try:
        # 실관람객 탭을 찾아 클릭
        tab = driver.find_element(By.LINK_TEXT, '실관람객')
        tab.click()
        time.sleep(2)  # 동적 콘텐츠 로드 대기
    except Exception as e:
        print(f"실관람객 탭 not found, skipping...")
        return None

    # 현재 페이지의 HTML 소스 가져오기
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    data = {
        'movie_name': movie['name'],  # URL 대신 영화 이름을 저장
    }

    # 실관람객 탭의 데이터 긁어오기
    data['real_audience_rating'] = soup.find('span', class_='area_star_number').text.strip() if soup.find('span', class_='area_star_number') else 'N/A'
    data['real_audience_part'] = soup.find('span', class_='area_people').text.strip() if soup.find('span', class_='area_people') else 'N/A'

    # 남성 비율
    man_section = soup.find('span', text='남자').find_next('span', class_='area_star_number') if soup.find('span', text='남자') else None
    data['real_audience_man_rating'] = man_section.text.strip() if man_section else 'N/A'

    # 여성 비율
    woman_section = soup.find('span', text='여자').find_next('span', class_='area_star_number') if soup.find('span', text='여자') else None
    data['real_audience_woman_rating'] = woman_section.text.strip() if woman_section else 'N/A'

    # 평점별 비율 추출
    nine_ten_section = soup.find('span', text='9-10')
    data['nine_ten_ratio'] = nine_ten_section.find_next('div', class_='area_text_percent').text.strip() if nine_ten_section else 'N/A'

    seven_eight_section = soup.find('span', text='7-8')
    data['seven_eight_ratio'] = seven_eight_section.find_next('div', class_='area_text_percent').text.strip() if seven_eight_section else 'N/A'

    five_six_section = soup.find('span', text='5-6')
    data['five_six_ratio'] = five_six_section.find_next('div', class_='area_text_percent').text.strip() if five_six_section else 'N/A'

    three_four_section = soup.find('span', text='3-4')
    data['three_four_ratio'] = three_four_section.find_next('div', class_='area_text_percent').text.strip() if three_four_section else 'N/A'

    first_two_section = soup.find('span', text='1-2')
    data['first_two_ratio'] = first_two_section.find_next('div', class_='area_text_percent').text.strip() if first_two_section else 'N/A'

    # 남자 비율 추출
    man_section = soup.find('li', class_='type_male')
    man_ratio = man_section.find('em', class_='num').text.strip() if man_section else 'N/A'
    data['man_ratio'] = man_ratio

    # 여자 비율 추출
    woman_section = soup.find('li', class_='type_female')
    woman_ratio = woman_section.find('em', class_='num').text.strip() if woman_section else 'N/A'
    data['woman_ratio'] = woman_ratio

    return data

data = []

# 각 영화의 실관람객 탭에 대한 데이터 수집
for movie in movie_urls:
    print(f"Processing movie: {movie['name']}")
    tab_data = fetch_data_from_tab(movie)
    if tab_data:
        data.append(tab_data)
    else:
        print(f"No data found for 실관람객 탭 in {movie['name']}")

driver.quit()  # 웹 드라이버 종료

# 데이터를 DataFrame으로 변환
df = pd.DataFrame(data)
print(df)

# DataFrame을 CSV로 저장
df.to_csv('audience_summary_data.csv', index=False, encoding='utf-8-sig')
```

```python
# 네티즌 요약 데이터 크롤링 코드

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import pandas as pd

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument('--headless')  # GUI 없이 실행
chrome_options.add_argument('--no-sandbox')  # 샌드박스 모드 비활성화
chrome_options.add_argument('--disable-dev-shm-usage')  # /dev/shm 파티션 사용 안함

# 웹 드라이버 설정 및 초기화
driver = webdriver.Chrome(options=chrome_options)

# 영화 URL 목록
movie_urls = [
    {"name": "패션왕(2014)", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1842967&qvt=0&query=%EC%98%81%ED%99%94%20%ED%8C%A8%EC%85%98%EC%99%95%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "신과함께 죄와벌", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=1819142&qvt=0&query=%EC%8B%A0%EA%B3%BC%ED%95%A8%EA%BB%98-%EC%A3%84%EC%99%80%20%EB%B2%8C%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "치즈인더트랩(2018)", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=4032538&qvt=0&query=%EC%98%81%ED%99%94%20%EC%B9%98%EC%A6%88%EC%9D%B8%EB%8D%94%ED%8A%B8%EB%9E%A9%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "여중생a", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%97%AC%EC%A4%91%EC%83%9Da+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%B9%98%EC%A6%88%EC%9D%B8%EB%8D%94%ED%8A%B8%EB%9E%A9+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdAgsqVOswsshsmHcossssstdZ-035695"},
    {"name": "신과함께 인과연", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%8B%A0%EA%B3%BC%ED%95%A8%EA%BB%98+%EC%9D%B8%EA%B3%BC%EC%97%B0+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%97%AC%EC%A4%91%EC%83%9Da+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdAUlqVN8VssbTT4OZssssstYo-472566"},
    {"name": "아이들은 즐겁다", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%95%84%EC%9D%B4%EB%93%A4%EC%9D%80+%EC%A6%90%EA%B2%81%EB%8B%A4+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%8B%A0%EA%B3%BC%ED%95%A8%EA%BB%98+%EC%9D%B8%EA%B3%BC%EC%97%B0+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdAPsqVN8osscGq6gsssssssmo-223076"},
    {"name": "천박사", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%B2%9C%EB%B0%95%EC%82%AC+%ED%87%B4%EB%A7%88+%EC%97%B0%EA%B5%AC%EC%86%8C+%EC%84%A4%EA%B2%BD%EC%9D%98+%EB%B9%84%EB%B0%80+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%95%84%EC%9D%B4%EB%93%A4%EC%9D%80+%EC%A6%90%EA%B2%81%EB%8B%A4+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdAFdqVOZCssRudTyZsssssti4-370972"},
    {"name": "모럴센스", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EB%AA%A8%EB%9F%B4%EC%84%BC%EC%8A%A4+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%B2%9C%EB%B0%95%EC%82%AC+%ED%87%B4%EB%A7%88+%EC%97%B0%EA%B5%AC%EC%86%8C+%EC%84%A4%EA%B2%BD%EC%9D%98+%EB%B9%84%EB%B0%80+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdAVsqVOsCss5ROM7CssssstYZ-155626"},
    {"name": "그녀의 버킷리스트", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EA%B7%B8%EB%85%80%EC%9D%98+%EB%B2%84%ED%82%B7%EB%A6%AC%EC%8A%A4%ED%8A%B8+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EB%AA%A8%EB%9F%B4%EC%84%BC%EC%8A%A4+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdAxsqVOsVss5khR0ossssssos-353610"},
    {"name": "반드시 잡는다", "url": "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=4146980&qvt=0&query=%EB%B0%98%EB%93%9C%EC%8B%9C%20%EC%9E%A1%EB%8A%94%EB%8B%A4%20%EA%B4%80%EB%9E%8C%ED%8F%89"},
    {"name": "바보", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EB%B0%94%EB%B3%B4+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EB%B0%94%EB%B3%B4+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBMspzL8VssnipdQKssssstVo-133955"},
    {"name": "순정만화", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%88%9C%EC%A0%95%EB%A7%8C%ED%99%94+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EB%B0%94%EB%B3%B4+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBewqVN8wss5TibQRssssst9R-065671"},
    {"name": "이끼", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%9D%B4%EB%81%BC+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%88%9C%EC%A0%95%EB%A7%8C%ED%99%94+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBTwqVOZCsseJhLjCssssssqV-326170"},
    {"name": "그대를 사랑합니다", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EA%B7%B8%EB%8C%80%EB%A5%BC+%EC%82%AC%EB%9E%91%ED%95%A9%EB%8B%88%EB%8B%A4+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%9D%B4%EB%81%BC+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBksqVN8wss4lSTtossssss%2FR-383953"},
    {"name": "이웃사람", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%9D%B4%EC%9B%83%EC%82%AC%EB%9E%8C+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EA%B7%B8%EB%8C%80%EB%A5%BC+%EC%82%AC%EB%9E%91%ED%95%A9%EB%8B%88%EB%8B%A4+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBQlqVOsCss6DA7QGssssstyw-222363"},
    {"name": "26년", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+26%EB%85%84+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%9D%B4%EC%9B%83%EC%82%AC%EB%9E%8C+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBAwqVOsoss6OM5lNssssst0h-295609"},
    {"name": "전설의주먹", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%A0%84%EC%84%A4%EC%9D%98%EC%A3%BC%EB%A8%B9+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+26%EB%85%84+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBEsqVOZosse1kr1wssssssU4-005508"},
    {"name": "은밀하게 위대하게", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%9D%80%EB%B0%80%ED%95%98%EA%B2%8C+%EC%9C%84%EB%8C%80%ED%95%98%EA%B2%8C+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%A0%84%EC%84%A4%EC%9D%98%EC%A3%BC%EB%A8%B9+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBrwqVOsoss5hXMmZssssssyV-258367"},
    {"name": "강철비", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EA%B0%95%EC%B2%A0%EB%B9%841+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%9D%80%EB%B0%80%ED%95%98%EA%B2%8C+%EC%9C%84%EB%8C%80%ED%95%98%EA%B2%8C+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBYlqVOZCssgdzmDossssssJV-348520"},
    {"name": "0.0MHz", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+0.0MHz+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EA%B0%95%EC%B2%A0%EB%B9%841+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBzwqVOZCssf4POiNssssstiZ-304250"},
    {"name": "발광하는 현대사", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EB%B0%9C%EA%B4%91%ED%95%98%EB%8A%94+%ED%98%84%EB%8C%80%EC%82%AC+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+0.0MHz+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdBHlqVOswsshc2NeRssssssZR-477597"},
    {"name": "시동", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%8B%9C%EB%8F%99+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EB%B0%9C%EA%B4%91%ED%95%98%EB%8A%94+%ED%98%84%EB%8C%80%EC%82%AC+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdB9wqVOswssTF6h7CssssssWC-349733"},
    {"name": "해치지않아", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%ED%95%B4%EC%B9%98%EC%A7%80%EC%95%8A%EC%95%84+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%8B%9C%EB%8F%99+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdCsspzLiwssnskzYZssssssX4-196857"},
    {"name": "강철비2", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EA%B0%95%EC%B2%A0%EB%B9%842+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%ED%95%B4%EC%B9%98%EC%A7%80%EC%95%8A%EC%95%84+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdCvwqVN8oss5KGnYGssssst64-482202"},
    {"name": "샤크:더 비기닝", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%83%A4%ED%81%AC+%EB%8D%94+%EB%B9%84%EA%B8%B0%EB%8B%9D+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EA%B0%95%EC%B2%A0%EB%B9%842+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=iwdCbspzLiwssdb44lossssstUs-477852"}
]


def fetch_data_from_tab(movie):
    driver.get(movie['url'])
    time.sleep(2)  # 웹 페이지 로드 대기

    try:
        # 네티즌 탭을 찾아 클릭
        tab = driver.find_element(By.LINK_TEXT, '네티즌')
        tab.click()
        time.sleep(2)  # 동적 콘텐츠 로드 대기
    except Exception as e:
        print(f"네티즌 탭 not found, skipping...")
        return None

    # 현재 페이지의 HTML 소스 가져오기
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    data = {
        'movie_name': movie['name'],  # URL 대신 영화 이름을 저장
    }

    # 네티즌 탭의 데이터 긁어오기
    data['netizen_rating'] = soup.find('span', class_='area_star_number').text.strip() if soup.find('span', class_='area_star_number') else 'N/A'
    data['netizen_part'] = soup.find('span', class_='area_people').text.strip() if soup.find('span', class_='area_people') else 'N/A'

    # 남성 비율
    man_section = soup.find('span', text='남자').find_next('span', class_='area_star_number') if soup.find('span', text='남자') else None
    data['netizen_man_rating'] = man_section.text.strip() if man_section else 'N/A'

    # 여성 비율
    woman_section = soup.find('span', text='여자').find_next('span', class_='area_star_number') if soup.find('span', text='여자') else None
    data['netizen_woman_rating'] = woman_section.text.strip() if woman_section else 'N/A'

    # 평점별 비율 추출
    nine_ten_section = soup.find('span', text='9-10')
    data['nine_ten_ratio'] = nine_ten_section.find_next('div', class_='area_text_percent').text.strip() if nine_ten_section else 'N/A'

    seven_eight_section = soup.find('span', text='7-8')
    data['seven_eight_ratio'] = seven_eight_section.find_next('div', class_='area_text_percent').text.strip() if seven_eight_section else 'N/A'

    five_six_section = soup.find('span', text='5-6')
    data['five_six_ratio'] = five_six_section.find_next('div', class_='area_text_percent').text.strip() if five_six_section else 'N/A'

    three_four_section = soup.find('span', text='3-4')
    data['three_four_ratio'] = three_four_section.find_next('div', class_='area_text_percent').text.strip() if three_four_section else 'N/A'

    first_two_section = soup.find('span', text='1-2')
    data['first_two_ratio'] = first_two_section.find_next('div', class_='area_text_percent').text.strip() if first_two_section else 'N/A'

    # 남자 비율 추출
    man_section = soup.find('li', class_='type_male')
    man_ratio = man_section.find('em', class_='num').text.strip() if man_section else 'N/A'
    data['man_ratio'] = man_ratio

    # 여자 비율 추출
    woman_section = soup.find('li', class_='type_female')
    woman_ratio = woman_section.find('em', class_='num').text.strip() if woman_section else 'N/A'
    data['woman_ratio'] = woman_ratio

    return data

data = []

# 각 영화의 네티즌 탭에 대한 데이터 수집
for movie in movie_urls:
    print(f"Processing movie: {movie['name']}")
    tab_data = fetch_data_from_tab(movie)
    if tab_data:
        data.append(tab_data)
    else:
        print(f"No data found for 네티즌 탭 in {movie['name']}")

driver.quit()  # 웹 드라이버 종료

# 데이터를 DataFrame으로 변환
df = pd.DataFrame(data)
print(df)

# DataFrame을 CSV로 저장
df.to_csv('netizen_summary_data.csv', index=False, encoding='utf-8-sig')
```

```python
import time
from os import write

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.common.actions.wheel_input import ScrollOrigin
import pymysql

# 자신의 크롬 버전에 맞는 chromedriver 설치
driver = webdriver.Chrome()
# 영화 '비상선언'
driver.get('https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=10105426&qvt=0&query=%EC%98%81%ED%99%94%20%EB%B9%84%EC%83%81%EC%84%A0%EC%96%B8%20%EC%A0%95%EB%B3%B4')
# 개봉일 부분 찾기
raw_released_date = driver.find_elements(By.XPATH, '/html/body/div[3]/div[2]/div/div[1]/div[2]/div[2]/div[1]/div/div[1]/dl/div[1]/dd')
released_date = raw_released_date[0].text
# 평점 창으로 이동
driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div/div[1]/div[2]/div[1]/div[3]/div/div/ul/li[4]/a').click()
time.sleep(0.5)
# 영화 제목
raw_movie_title = driver.find_elements(By.XPATH, '/html/body/div[3]/div[2]/div/div[1]/div[2]/div[1]/div[1]/h2/span/strong')
movie_title = raw_movie_title[0].text
# 웹페이지 스크롤 내리기 (y축 방향으로 +600)
ActionChains(driver) \
        .scroll_by_amount(0, 600) \
        .perform()
# 아래 코드 진행을 위해 잠시 딜레이시키기
time.sleep(2)
# 스포일러 감상 보여 주기 버튼 클릭
driver.find_element(By.XPATH, "/html/body/div[3]/div[2]/div/div[1]/div[2]/div[2]/div[2]/div/div[2]/div[2]/div/div[2]/button").click()
# 영화 리뷰에 대한 인덱스
comment_index = 1
# 리뷰가 공백인 경우 카운트
nonsave_number = 0
# pymysql 연결
mydb = pymysql.connect(host='localhost', user='root', password='personal pw', db='personal database name', charset='utf8')
mycursor = mydb.cursor()
# insert sql query
sql = "INSERT INTO reviewInformation(movie_title, comment, star_score, upvote, downvote, writing_date, released_date) VALUES (%s, %s, %s, %s, %s, %s, %s)"
# 영화 제목, 개봉일 출력
print(movie_title)
print(released_date)
time.sleep(1)
while True:
    # 관람객 코멘트 부분 찾기
    raw_comment = driver.find_elements(By.XPATH,'/html/body/div[3]/div[2]/div/div[1]/div[2]/div[2]/div[2]/div/div[2]/div[6]/ul/li[' + str(comment_index) + ']/div[2]/div/span[2]')
    # 크롤링 종료
    if len(raw_comment) == 0:
        break
    comment = raw_comment[0].text
    # 관람객 별점 부분 찾기
    raw_star_score = driver.find_elements(By.XPATH, '/html/body/div[3]/div[2]/div/div[1]/div[2]/div[2]/div[2]/div/div[2]/div[6]/ul/li[' + str(comment_index) + ']/div[1]/div/div[2]')
    star_score = raw_star_score[0].text[13:]
    # 리뷰 작성일 부분 찾기
    raw_writing_date = driver.find_elements(By.XPATH, '/html/body/div[3]/div[2]/div/div[1]/div[2]/div[2]/div[2]/div/div[2]/div[6]/ul/li[' + str(comment_index) + ']/dl/dd[2]')
    writing_date = raw_writing_date[0].text[0:11]
    # 공감수 부분 찾기
    raw_upvote = driver.find_elements(By.XPATH, '/html/body/div[3]/div[2]/div/div[1]/div[2]/div[2]/div[2]/div/div[2]/div[6]/ul/li[' + str(comment_index) + ']/div[3]/button[1]/span')
    upvote = raw_upvote[0].text
    # 비공감수 부분 찾기
    raw_downvote = driver.find_elements(By.XPATH, '/html/body/div[3]/div[2]/div/div[1]/div[2]/div[2]/div[2]/div/div[2]/div[6]/ul/li[' + str(comment_index) + ']/div[3]/button[2]/span')
    downvote = raw_downvote[0].text
    time.sleep(0.5)
    if len(comment) > 0:
        print("================================================================================================")
        print("작성일:", writing_date)    # 작성일 출력
        print("별점:", star_score) # 별점 출력
        print("감상평:", comment)  # 코멘트 출력
        print("공감수:", upvote, end=' ')  # 공감수 출력
        print("비공감수:", downvote)    # 비공감수 출력
        print("리뷰 총합:", comment_index - nonsave_number)
        # sql query 문 적용
        mycursor.execute(sql, (movie_title, comment, star_score, upvote, downvote, writing_date, released_date))
        mydb.commit()
    else:
        nonsave_number += 1
    comment_index += 1
    # 스크롤 내리기 위해 영화 평점 창 쪽으로 초점 이동
    scroll_origin = ScrollOrigin.from_viewport(140, 641)
    # 스크롤 내리기
    ActionChains(driver) \
        .scroll_from_origin(scroll_origin, 0, 185) \
        .perform()
exit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--no-sandbox")
options.add_argument("--disable-gpu")
options.add_argument("--headless")
options.add_argument("--disable-software-rasterizer")
options.add_argument('--disable-extensions')
options.add_argument("--remote-debugging-port=9222")
options.add_argument("--verbose")
options.add_argument("--log-path=/path/to/chromedriver.log")

driver = webdriver.Chrome(options=options)

def fetch_reviews_from_page(movie):
    driver.get(movie['url'])
    time.sleep(2)  # Allow page to load

    # Scroll within the specific div that has its own scrollbar
    scroll_container = driver.find_element(By.CSS_SELECTOR, '.lego_review_list._scroller')
    last_height = driver.execute_script("return arguments[0].scrollHeight", scroll_container)

    while True:
        # Execute script to scroll within the specified div
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll_container)
        time.sleep(3)  # Wait for the page to load more items

        new_height = driver.execute_script("return arguments[0].scrollHeight", scroll_container)
        if new_height == last_height:
            break  # Break the loop when no more data is loaded
        last_height = new_height

    # Now, extract the reviews
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    reviews = []
    comments = soup.find_all('li', class_='area_card _item')
    for comment in comments:
        if comment.find('span', text='관람객'):  # Check if the review is from an actual viewer
            review_text = comment.find('span', class_='desc _text').get_text(strip=True)
            star_score = comment.find('div', class_='area_text_box').get_text(strip=True)
            like_count = comment.select_one('.area_button_upvote ._count_num').get_text(strip=True)
            dislike_count = comment.select_one('.area_button_downvote ._count_num').get_text(strip=True)

            reviews.append({
                'movie_name': movie['name'],
                'review_text': review_text,
                'star_score': star_score,
                'like_count': like_count,
                'dislike_count': dislike_count,
            })

    return reviews

movie_urls = [
    {"name": "옥수역귀신", "url": "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EC%98%81%ED%99%94+%EC%98%A5%EC%88%98%EC%97%AD%EA%B7%80%EC%8B%A0+%EA%B4%80%EB%9E%8C%ED%8F%89&oquery=%EC%98%81%ED%99%94+%EC%95%84%EB%B6%80%EC%9F%81%EC%9D%B4+%EA%B4%80%EB%9E%8C%ED%8F%89&tqi=ixZFLwqo1e8ssBKJKXKssssssis-427824"}
]

all_reviews = []
for movie in movie_urls:
    reviews = fetch_reviews_from_page(movie)
    all_reviews.extend(reviews)
    print(f"Collected {len(reviews)} reviews for {movie['name']} from actual viewers.")

driver.quit()

# Convert all_reviews to DataFrame and save to CSV
df = pd.DataFrame(all_reviews, columns=['movie_name', 'review_text', 'star_score', 'like_count', 'dislike_count'])
df.to_csv('brand_new_audience_reviews.csv', index=False, encoding='utf-8-sig')
```
