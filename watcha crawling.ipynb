```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
from selenium.webdriver.common.keys import Keys
import csv

# Chrome 옵션 설정 (백그라운드에서 실행되도록)
chrome_options = Options()
chrome_options.add_argument("--headless")  # 창이 안 보이도록 실행
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# ChromeDriver 경로 설정
chrome_driver_path =

# 드라이버 설정
driver = webdriver.Chrome(executable_path=chrome_driver_path, options=chrome_options)

# 드라마 이름 리스트
movie_names = [
    "이미테이션", "그림자 미녀", "미생"  # 예시 드라마 리스트
]

# 페이지가 로드될 시간을 충분히 기다림
time.sleep(5)

# 댓글 내용 저장 리스트
all_comments_data = []

# 각 영화에 대한 댓글을 수집하는 함수
def collect_drama_comments(movie_name, max_scrolls=50):
    # 드라마 이름으로 검색된 URL로 이동
    search_url = f"https://pedia.watcha.com/ko-KR/search?query={movie_name}"
    driver.get(search_url)

    # 페이지 로딩 시간
    time.sleep(3)

    try:
        # 검색 결과에서 첫 번째 드라마 클릭 (적절한 요소를 찾아 클릭하도록 수정 가능)
        first_result = driver.find_element(By.CSS_SELECTOR, 'a[href^="/ko-KR/contents/"]')
        drama_url = first_result.get_attribute('href') + "/comments"  # /comments 추가
        driver.get(drama_url)
        time.sleep(3)
    except Exception as e:
        print(f"Error finding the drama page for {drama_name}: {e}")
        return

    comments_data = []
    last_height = driver.execute_script("return document.body.scrollHeight")
    scroll_count = 0

    while scroll_count < max_scrolls:
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(5)

        try:
            comment_elements = driver.find_elements(By.CSS_SELECTOR, 'a[href^="/ko-KR/comments/"]')

            for comment_element in comment_elements:
                text_divs = comment_element.find_elements(By.TAG_NAME, 'div')
                for div in text_divs:
                    if div.text.strip():
                        comment_text = div.text
                        if comment_text not in comments_data:
                            comments_data.append(comment_text)

            print(f"{drama_name} - 스크롤 {scroll_count + 1}/{max_scrolls} 완료. 현재까지 수집된 댓글: {len(comments_data)}개")

        except Exception as e:
            print(f"Error: {e}")
            break

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            print(f"{drama_name} - 더 이상 스크롤할 내용이 없습니다.")
            break

        last_height = new_height
        scroll_count += 1

    all_comments_data.extend([(drama_name, comment) for comment in comments_data])  # 드라마 이름과 댓글 저장

# 드라마 목록을 돌며 댓글 수집
for drama in drama_names:
    collect_drama_comments(drama)

# 수집된 모든 댓글 출력
for idx, (drama, comment) in enumerate(all_comments_data):
    print(f"{idx + 1}. {drama} - 댓글 내용: {comment}")

# 드라이버 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd

# 웹드라이버 설정 (크롬 예시)
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 브라우저 창을 띄우지 않음
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')


driver = webdriver.Chrome(options=options)

# 크롤링할 URL 목록 (여기에 원하는 만큼 추가 가능)
movie_urls = [
    {"name": "패션왕(2014)", "url": "https://pedia.watcha.com/ko-KR/contents/m5NV4Y8/comments"}
]


# 크롤링 데이터를 저장할 리스트 준비
all_data = []

for movie in movie_urls:
    # 웹사이트 접속
    driver.get(movie['url'])  # movie 딕셔너리에서 URL 값을 추출하여 전달

    time.sleep(2)  # 페이지 로딩 대기

    # 팝업 닫기 (필요시 팝업 처리)
    try:
        popup_button = driver.find_element(By.CSS_SELECTOR, 'button.hsDVweTz')
        popup_button.click()
        time.sleep(1)  # 버튼 클릭 후 대기
    except:
        print("팝업이 나타나지 않았습니다. 계속 진행합니다.")

    # 데이터 수집할 리스트 준비
    title = ''
    likes_list = []
    star_ratings = []
    comments = []

    try:
        # 제목 수집
        title = driver.find_element(By.CSS_SELECTOR, 'h1.QCY16Srr').text

        # 댓글 페이지로 이동
        comments_button = driver.find_element(By.CSS_SELECTOR, 'a[href*="/comments"]')
        comments_button.click()

        time.sleep(2)  # 페이지 로딩 대기

        # 스크롤 전 페이지 높이
        last_height = driver.execute_script("return document.body.scrollHeight")

        total_comment_limit = 1000  # 수집할 댓글의 최대 개수
        max_scroll_count = 100  # 스크롤을 제한하는 횟수 (필요시)
        scroll_count = 0

        while True:
            # 현재 페이지의 소스를 가져와 BeautifulSoup으로 파싱
            soup = BeautifulSoup(driver.page_source, 'html.parser')

            # 댓글, 좋아요, 별점 요소를 선택하여 수집
            comment_elements = soup.select('div.eqSewv3p.B79Ex5Ar')
            like_elements = soup.select('div.TmaB244V.Zcvzhajf em')  # 좋아요 <em> 태그
            star_elements = soup.select('div.aytsxOVO.Zcvzhajf span')  # 별점 <span> 태그

            # 댓글 수집
            for comment in comment_elements:
                if len(comments) >= total_comment_limit:
                    break  # 댓글 1000개 수집하면 멈춤
                comments.append(comment.text)

            # 좋아요 수집
            for like in like_elements:
                likes_list.append(like.text)

            # 별점 수집
            for star in star_elements:
                star_ratings.append(star.text)

            # 더 이상 새로운 댓글이 없거나 1000개 수집 완료 시 종료
            if len(comments) >= total_comment_limit:
                print(f"최대 댓글 수({total_comment_limit})에 도달했습니다. 크롤링을 종료합니다.")
                break

            # 스크롤 내리기
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

            # 스크롤 후 페이지 높이 확인
            new_height = driver.execute_script("return document.body.scrollHeight")

            # 더 이상 스크롤이 되지 않으면 종료
            if new_height == last_height:
                print("더 이상 스크롤이 내려가지 않습니다. 크롤링을 종료합니다.")
                break

            # 페이지 높이 업데이트
            last_height = new_height
            scroll_count += 1

            # 스크롤 횟수 제한
            if scroll_count >= max_scroll_count:
                print("최대 스크롤 횟수에 도달했습니다. 크롤링을 종료합니다.")
                break

    except Exception as e:
        print(f"크롤링 중 오류 발생: {e}")

    finally:
        # 각 요소들의 길이가 동일한지 확인 후, 불일치 시 데이터를 맞춤
        min_len = min(len(comments), len(likes_list), len(star_ratings))
        comments = comments[:min_len]
        likes_list = likes_list[:min_len]
        star_ratings = star_ratings[:min_len]

        # 크롤링한 데이터를 리스트에 추가
        for i in range(min_len):
            all_data.append({
                'Title': title,
                'Likes': likes_list[i],
                'Star Rating': star_ratings[i],
                'Comments': comments[i]
            })

# 모든 데이터를 하나의 DataFrame으로 저장
df = pd.DataFrame(all_data)

# CSV로 저장
df.to_csv('watcha_movie_reviews.csv', index=False)

print("CSV 파일이 생성되었습니다: watcha_movie_r.csv")

# 브라우저 종료
driver.quit()
```
